```python
#Import games from Regular Season Data

import csv

games_data=[]

with open('RegularSeasonDetailedResults.csv') as csvfile:
    reader=csv.DictReader(csvfile)
    for row in reader:
        game=row
        games_data.append(game)
```

```python
#Create ordered dictionaries of team averages
#Each team has an assigned team ID in the form of a number

from collections import OrderedDict
import copy

key_array=['points','fgm', 'fga', 'fgm3', 'fga3', 'ftm', \
           'fta', 'or', 'dr', 'ast', 'to', 'stl', 'blk', 'pf', 'SOS', 'AdjO', 'AdjD', 'OppO']

team_data=OrderedDict()
default_dict=OrderedDict()

for i in range(2003,2018):
    default_dict[str(i)]=OrderedDict()
    default_dict[str(i)]['gp']=0
    default_dict[str(i)]['gpcount']=0
    default_dict[str(i)]['last12']=0
    for j in key_array:
        default_dict[str(i)][j]=0

with open('Teams.csv') as csvfile:
    reader=csv.DictReader(csvfile)
    for row in reader:
        team=row['Team_Id']
        team_data[team]=copy.deepcopy(default_dict)
```

```python
#Taking totals: 
#Divide by possessions during this process....

key_array2=['Wscore','Wfgm', 'Wfga', 'Wfgm3', 'Wfga3', 'Wftm', \
           'Wfta', 'Wor', 'Wdr', 'Wast', 'Wto', 'Wstl', 'Wblk', 'Wpf','Lscore','Lfgm', \
           'Lfga', 'Lfgm3', 'Lfga3', 'Lftm', 'Lfta', 'Lor', 'Ldr', 'Last', 'Lto', 'Lstl', 'Lblk', 'Lpf']

for game in games_data:
    team_data[game['Wteam']][game['Season']]['gp']=team_data[game['Wteam']][game['Season']]['gp']+1
    team_data[game['Lteam']][game['Season']]['gp']=team_data[game['Lteam']][game['Season']]['gp']+1
    
    poss=0.5*(float(game['Wfga'])+0.475*float(game['Wfta'])-float(game['Wor'])+float(game['Wto']))+\
    0.5*(float(game['Lfga'])+0.475*float(game['Lfta'])-float(game['Lor'])+float(game['Lto']))
    
    for v in range(len(key_array2)):
        if(v<len(key_array2)/2):
            team_data[game['Wteam']][game['Season']][key_array[v]]=\
            team_data[game['Wteam']][game['Season']][key_array[v]]+float(game[key_array2[v]])/poss
        else:
            index=int(v-len(key_array2)/2)
            team_data[game['Lteam']][game['Season']][key_array[index]]=\
            team_data[game['Lteam']][game['Season']][key_array[index]]+float(game[key_array2[v]])/poss
            
    
        
```

```python
#Adjust data for average stat per game
for team in team_data:
    for season in team_data[team]:
        for stat in team_data[team][season]:
            if(stat!='gp'):
                if(team_data[team][season]['gp']!=0):
                    team_data[team][season][stat]=team_data[team][season][stat]/team_data[team][season]['gp']
```

```python
#Import & Add Strength of Schedule Data

import csv
import os


for i in range(2003,2018):
    script_dir = os.path.dirname('__file__') #<-- absolute dir the script is in
    rel_path = 'CSVsos/'+str(i)+'.csv'
    abs_file_path = os.path.join(script_dir, rel_path)
    with open(abs_file_path) as csvfile:
        reader=csv.DictReader(csvfile)
        for row in reader:
            game=row
            try:
                team_data[game['Team']][str(i)]['SOS']=float(game['SOS'])
                team_data[game['Team']][str(i)]['AdjO']=float(game['AdjO'])
                team_data[game['Team']][str(i)]['AdjD']=float(game['AdjD'])
                team_data[game['Team']][str(i)]['OppO']=float(game['OppO'])
            except:
                team_data[game['Team']][str(i)]['SOS']=0
```

```python
#Organize Last 12 Wins Data:

for game in games_data:
    team_data[game['Wteam']][game['Season']]['gpcount'] = team_data[game['Wteam']][game['Season']]['gpcount']+1
    team_data[game['Lteam']][game['Season']]['gpcount'] = team_data[game['Lteam']][game['Season']]['gpcount']+1
    
    if(team_data[game['Wteam']][game['Season']]['gpcount']>team_data[game['Wteam']][game['Season']]['gp']-12):
        team_data[game['Wteam']][game['Season']]['last12']=team_data[game['Wteam']][game['Season']]['last12']+1
```

```python
#Extract Features. Each Stat is a Feature

import numpy as np

def extract_features(data):
    xtrain = []
    for game in data:
        fvector = []
        for j in game.values():
            if(j=='N'):
                fvector.append(0)
            elif(j=='H'):
                fvector.append(1)
            elif(j=='A'):
                fvector.append(-1)
            else:
                fvector.append(j)
        
        fvector = np.asarray(fvector) #turns fvector into numpy array
        xtrain.append(fvector)

    return np.asarray(xtrain)
```
